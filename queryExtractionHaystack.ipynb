```python
def handleChildren(dict):
    t = dict['type']
    text = ''
    if t=='p':
        #handle array
        for i in dict['children']:
            if 'children' in i.keys():
                #recurse
                text += handleChildren(i)
                text+= ' '
            else:
                if 'text' in i.keys():
                    text += i['text']
                    text += ' '
    elif t=='table':
        #handle table
        for i in dict['children']:
            for j in i['children']:
                if 'text' in j['children'][0].keys():
                    text += j['children'][0]['text']
                    text += ' '
    elif t=='ul':
        #handle list
        for i in dict['children']:
            if 'text' in i['children'][0]['children'][0].keys():
                text += i['children'][0]['children'][0]['text']
                text += ' '
    elif t=='ol':
        #handle list
        for i in dict['children']:
            if 'text' in i['children'][0]['children'][0].keys():
                text += i['children'][0]['children'][0]['text']
                text += ' '
    elif t=='img':
        # ignore image for now
        pass
    else:
        # see if any other type
        if 'children' in dict.keys():
            for i in dict['children']:
                if 'children' in i.keys():
                    #recurse
                    text += handleChildren(i)
                    text+= ' '
                else:
                    if 'text' in i.keys():
                        text += i['text']
                        text += ' '
        elif 'text' in dict.keys():
            text += dict['text']
            text += ' '
                    
    return text
        
```

```python
k = 0
for i in Dfilter.index:
    try:

        t = ast.literal_eval(Dfilter['content'][i])
        #print(t)
        text = ''
        if 'editorContent' in t.keys():
            #editorContent
            for j in t['editorContent']:
                text += handleChildren(j) 
                new_row = {
                  'title':Dfilter['title'][i],
                  'department':Dfilter['department'][i],
                  'pid':k,
                  'text':text
                }
                newData = newData.append(new_row, ignore_index=True)
            #print(text)
        elif 'blocks' in t.keys():
            #blocks
            for j in t['blocks']:
                text += handleChildren(j)
                new_row = {
                  'title':Dfilter['title'][i],
                  'department':Dfilter['department'][i],
                  'pid':k,
                  'text':text
                }
                newData = newData.append(new_row, ignore_index=True)
            #print(text)
            
        
        k+=1
    except:
        pass
```

```python
launch_es()
document_store = ElasticsearchDocumentStore()
document_store.delete_documents()
document_store.write_documents(got_dicts)

# Initialize Sparse retriever
es_retriever = ElasticsearchRetriever(document_store=document_store)

# Initialize dense retriever
dpr_retriever = DensePassageRetriever(document_store)
document_store.update_embeddings(dpr_retriever, update_existing_embeddings=False)

```

```python
reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2")

sklearn_keyword_classifier = Pipeline()
sklearn_keyword_classifier.add_node(component=SklearnQueryClassifier(), name="QueryClassifier", inputs=["Query"])
sklearn_keyword_classifier.add_node(component=dpr_retriever, name="DPRRetriever", inputs=["QueryClassifier.output_1"])
sklearn_keyword_classifier.add_node(component=es_retriever, name="ESRetriever", inputs=["QueryClassifier.output_2"])
sklearn_keyword_classifier.add_node(component=reader, name="QAReader", inputs=["ESRetriever", "DPRRetriever"])
sklearn_keyword_classifier.draw("pipeline_classifier.png")
```
